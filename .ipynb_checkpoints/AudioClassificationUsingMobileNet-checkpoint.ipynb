```python
def load_wav_16k_mono(path):
    file_contents = tf.io.read_file(path)
    wav, sample_rate = tf.audio.decode_wav(file_contents, desired_channels=1)
    wav = tf.squeeze(wav, axis=-1)
    sample_rate = tf.cast(sample_rate, dtype=tf.int64)
    wav = tfio.audio.resample(wav, rate_in=sample_rate, rate_out=16000)
    return wav
```

## Create Tensorflow Datasets

```python
data_path = tf.data.Dataset.list_files(base_dir + '/*.wav')
def extract_class(file_path):
    file_name = tf.strings.split(file_path, '/')[-1]
    class_name = tf.strings.split(file_name, '_')[0]
    return file_path, int(class_name)
dataset = data_path.map(lambda x: extract_class(x))
```

```python
data_path_test = tf.data.Dataset.list_files(base_dir_test + '/*.wav')
```

```python
dataset.shuffle(1000).as_numpy_iterator().next()
```

```python
def preprocess(file_path, label):
    wav = load_wav_16k_mono(file_path)
    wav = wav[:48000]
    zero_padding = tf.zeros([48000] - tf.shape(wav), dtype=tf.float32)
    wav = tf.concat([zero_padding, wav], 0)
    spectogram = tf.signal.stft(wav, frame_length=1024, frame_step=256)
    spectogram = tf.abs(spectogram)
    spectogram = tf.expand_dims(spectogram, axis=2)
    label = tf.convert_to_tensor(label)
    label = tf.one_hot(label, depth=10)
    return spectogram, label
```

```python
filepath, label = dataset.shuffle(buffer_size=1000).as_numpy_iterator().next()
```

```python
spectogram, label = preprocess(filepath, label)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(spectogram)
plt.show()
```

```python
train = data_path.map(lambda x: extract_class(x))
train = train.map(preprocess)
train = train.shuffle(buffer_size=1000)
train = train.batch(16)
train = train.prefetch(8)
```

```python
test = data_path_test.map(lambda x: extract_class(x))
test = test.map(preprocess)
test = test.batch(16)
test = test.prefetch(8)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Activation
from keras.layers import Flatten
from keras.layers import Dense, GlobalAveragePooling1D
from tensorflow.keras import layers, models, metrics

model = Sequential()
model.add(layers.Input(shape=(batch[0].shape[1], batch[0].shape[2], 1)))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(layers.BatchNormalization())
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))
model.add(layers.BatchNormalization())
model.add(layers.Dropout(0.2))
model.add(MaxPooling2D(2, 2))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(layers.BatchNormalization())
model.add(MaxPooling2D(2, 2))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(layers.Dropout(0.4))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', metrics.Precision(), metrics.Recall()])
model.summary()
```

```python
hist = model.fit(train, epochs=30, validation_data=test)
```

```python
plt.title('Loss')
plt.plot(hist.history[list(hist.history.keys())[0]], 'r')
plt.plot(hist.history[list(hist.history.keys())[3]], 'b')
plt.show()
```

```python
from pydub import AudioSegment
def convert_sound(input_file: str, output_file: str, output_format: str):
    audio = AudioSegment.from_file(input_file)
    audio.export(output_file, format=output_format)
    print(f"File converted and saved as {output_file}")
def convert_music_format(input_file, output_file):
    output_format = output_file.split(os.sep)[-1].split('.')[-1]
    convert_sound(input_file, output_file, output_format)
```

```python
input_path = "test1.m4a"
output_path = "output4.wav"



if input_path.split('.')[-1] != 'wav':
    convert_music_format(input_path, output_path)
    x = output_path
else:
    x = input_path
```

```python
wav = load_wav_16k_mono(x)
wav = wav[:50000]
zero_padding = tf.zeros([50000] - tf.shape(wav), dtype=tf.float32)
wav = tf.concat([zero_padding, wav], 0)
spectogram = tf.signal.stft(wav, frame_length=320, frame_step=32)
spectogram = tf.abs(spectogram)
spectogram = tf.expand_dims(spectogram, axis=2)
spectogram = tf.expand_dims(spectogram, axis=0)
prediction = model(spectogram)
prediction.numpy().argmax()
```
